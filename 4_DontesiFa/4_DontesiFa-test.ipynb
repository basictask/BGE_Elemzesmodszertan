### Adatok ábrázolásához segéd metódus: ponthálóvá alakítás

```python
from dtreeviz.trees import dtreeviz
from sklearn import preprocessing

df = pd.read_csv('halak.csv', header=0, sep=';', encoding='ISO-8859-2')

cols = ['Hossz1', 'Magassag']

X = np.array(df[cols]) # Független változók
Y = df['Faj']

le = preprocessing.LabelEncoder()
y_le = le.fit_transform(Y)

clfx = DecisionTreeClassifier(random_state=42)

clfx.fit(X, y_le)

viz = dtreeviz(clfx, X, y_le,
                target_name="target",
                feature_names=cols,
                class_names=list(set(Y)))
```
